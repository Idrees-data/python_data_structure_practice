```python
def remove_dup(dup):
    result = []
    for i in dup:
        if i not in result:
            result.append(dup)
        return result
x = [1,2,2,3,1,4]
remove_dup(x)
```

```python
word = 'helloo'
freq = {}
for ch in word:
    if ch in freq:
        freq[ch]+=1
    else:
        freq[ch] = 1
print(freq)
```

```python
def counter_freq(text):
    freq = {}
    for ch in text:
        if ch in freq:
            freq[ch]+=1
        else:
            freq[ch]=1
    print(freq)
counter_freq('hello')
```

```python
def counter_freq_word(word):
    freq = {}
    for ch in word:
        if ch in freq:
            freq [ch]+=1
        else:
            freq[ch]=1
    print(freq)
sentence = "Oye,Morning Sunshine"
word = sentence.split()
counter_freq_word(word)
```
